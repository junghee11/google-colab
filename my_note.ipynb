```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace( -5, 5 )

y_1 = 2 *x
plt.plot(x,y_1)
plt.show()

y_2 = 3 * x

plt.xlabel( "x value" )
plt.ylabel( "y value" )
plt.title( "My Graph" )
plt.plot( x, y_1, label="y1" )
plt.plot( x, y_2, label="y2", linestyle="dashed")
plt.legend()

plt.show()
```
